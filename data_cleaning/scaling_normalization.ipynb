```python
import numpy as np 
original_data = np.random.exponential(size=100)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn
```

```python
original_data[:10]
```

```python
original_data = original_data.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
standard_scale = StandardScaler()
standard_scale.fit(original_data)
scaled_data = standard_scale.transform(original_data)
```

```python
min_max_scale = MinMaxScaler(feature_range=(-2,2))
min_max_scale_data= min_max_scale.fit_transform(original_data)
```

```python
print(scaled_data[:5])
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,3))
sns.histplot(original_data,ax=ax[0],kde=True,legend=False)
ax[0].set_title("Original Data")
sns.histplot(scaled_data,ax=ax[1],kde=True,legend=False)
ax[1].set_title("Scaled data")
sns.histplot(min_max_scale_data,ax=ax[2],kde=True,legend=False)
ax[2].set_title("Minmax data")
plt.show()
```

```python
# lets make a custom StandarScaler function 
```

```python
def standardscaler_fn(x):
    mean = np.mean(x,axis=0)
    std = np.std(x,axis=0,ddof=1)
    return (x-mean)/std
```

```python
a = standardscaler_fn(original_data)
```

```python
print(a[:5])
```

# Normalization

```python
from sklearn.preprocessing import normalize
```

```python
x_normalized = normalize(original_data,norm='l2',axis=0)
```

```python
fig,ax = plt.subplots(1,2,figsize=(15,3))
sns.histplot(original_data,legend=False,ax=ax[0],kde=True)
ax[0].set_title("Original Data")
sns.histplot(x_normalized,legend=False,ax=ax[1],kde=True)
ax[1].set_title("Normalize Data")
```
